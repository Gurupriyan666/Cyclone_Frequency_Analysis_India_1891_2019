# üìä Analysis of Cyclone & Depression Frequency in India (1891‚Äì2019)

## üìå Introduction

India is frequently affected by cyclones and tropical depressions due to its long coastline and geographical location. Understanding historical cyclone trends is important for disaster preparedness, climate research, and policy planning.

This project analyzes cyclone and depression frequency data from 1891 to 2019 using Python to identify long-term trends, seasonal patterns, and decade-wise variations.

## üéØ Objectives

* Analyze long-term cyclone trends in India

* Compare seasonal cyclone frequency

* Identify climate pattern changes

* Create visual insights using Python

## üìÇ Dataset Information

* Source: data.gov.in

* Time Period: 1891‚Äì2019

* Description: Annual and seasonal cyclone & depression frequency in India

* Format: Excel (.xls)

```python
### üì• Data Loading

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("seasonalfrequency_c_monsoon-1891-2019_final.xlsx")
df.head()
```

```python
### üîç Data Understanding
df.info()
df.describe()
```

```python
### üßπ Data Cleaning

df.columns = df.columns.str.strip().str.lower().str.replace(" ","_")
df.isnull().sum()
```

```python
df.fillna(0,inplace=True)
df.head
```

## üß† Feature Engineering

### Create Total Column


```python
season_columns = df.columns.drop('year')

season_columns
```

```python
df['total'] = df[season_columns].sum(axis=1)
df.head()
```

## üìà Yearly Trend Analysis

This section analyzes how cyclone and depression frequency has changed over time.

### üìà Yearly Trend Plot

```python
plt.figure()
plt.plot(df['year'], df['total'])
plt.title("Total Cyclones & Depressions Over Years")
plt.xlabel("Year")
plt.ylabel("Frequency")
plt.show()
```

### Yearly Trend Insights

- Cyclone and depression frequency shows high variability across years.
- Earlier decades display more frequent high peaks in cyclone activity.
- Post-1970 period shows comparatively lower average frequency with occasional spikes.
- The trend indicates strong climate variability rather than a consistent increase or decrease.
- Year-wise analysis is important for understanding extreme climate behavior.


## üå¶ Seasonal Trend Analysis

This compares cyclone frequency across seasons.

### üå¶ Seasonal Plot

```python
season_totals = [col for col in season_columns if 'total' in col]
season_totals
```

```python
df[season_totals].plot(kind='line')
plt.title("Season-wise Cyclone Frequency Trend (Totals Only)")
plt.xlabel("Year")
plt.ylabel("Frequency")
plt.show()
```

### Seasonal Trend Insights

- Cyclone frequency during the June‚ÄìSeptember combined season is consistently higher than individual monthly totals, showing that monsoon months contribute most to cyclone activity.

- June and July show relatively higher variability compared to August and September.

- Most seasonal cyclone counts remain between 0 and 3 events per year, indicating low-to-moderate frequency in individual months.

- Seasonal cyclone activity shows strong year-to-year fluctuations, highlighting climate-driven variability.

- No season shows a perfectly increasing or decreasing trend, suggesting cyclone patterns are irregular rather than linear.

- The combined seasonal total provides a clearer representation of cyclone risk compared to individual monthly values.



## üìä Seasonal Average Comparison

This identifies the most active cyclone season.

### üìä Seasonal Average Plot

```python
df[season_columns].mean().plot(kind= 'bar')
plt.title("Average Seasonal Cyclone Frequency")
plt.show()
```

### Seasonal Average Comparison Insights

- The combined June‚ÄìSeptember season shows the highest average cyclone frequency, indicating that most cyclone activity occurs during the monsoon period.

- Among individual months, September records a higher average cyclone frequency compared to June, July, and August.

- Land-based cyclone occurrences are consistently lower than sea-based occurrences, suggesting that most cyclones weaken before reaching land.

- The Arabian Sea and Bay of Bengal show different contribution levels, highlighting regional variation in cyclone formation.

- Individual monthly averages remain relatively low, while the combined seasonal average is significantly higher, emphasizing the importance of seasonal aggregation in


## üîÅ Decade-wise Analysis

This groups data by decades.

### üîÅ Decade Plot

```python
df['decade'] = (df['year']//10)*10

df.groupby('decade')['total'].mean().plot(kind= 'bar')
plt.title("Average Cyclone Frequency by Decade")
plt.show()
```

### Decade-wise Trend Insights

- The highest average cyclone frequency is observed in the late 19th and early 20th centuries, particularly during the 1890s and 1900s.

- A gradual decline in average cyclone frequency is visible from the 1940s to the 1960s.

- The 1970s show a noticeable increase compared to the previous decade, indicating a temporary rise in cyclone activity.

- From the 1980s onward, cyclone frequency drops significantly and remains relatively low through the 2000s and 2010s.

- The overall long-term pattern suggests a declining trend in average cyclone frequency across decades, with intermittent fluctuations.

- This decade-wise analysis highlights that cyclone activity is not constant and is influenced by long-term climatic variations.
